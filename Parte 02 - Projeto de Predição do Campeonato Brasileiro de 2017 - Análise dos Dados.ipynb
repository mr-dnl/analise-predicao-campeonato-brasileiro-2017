<h1 align="center"> Análide dos Dados Carregados<h/1>

```python
# Importação das bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set(style="darkgrid")
```

```python
#Leitura do csv
df_campeonato = pd.read_csv('df_campeonato.csv',sep=';', encoding='utf-8')
```

```python
df_campeonato.head()
```

```python
# Total de partidas
tt_partidas = df_campeonato.shape[0]

# Partidas vencidadas pelo time da casa
tt_vt_tm_cs = len(df_campeonato.query('FinalCasa == "v"'))
tt_em_tm_cs = len(df_campeonato.query('FinalCasa == "e"'))
tt_dr_tm_cs = len(df_campeonato.query('FinalCasa == "d"'))

# Porcentagem de vitórias
pc_vt_cs = (float(tt_vt_tm_cs) / (tt_partidas)) * 100
pc_em_cs = (float(tt_em_tm_cs) / (tt_partidas)) * 100
pc_dr_cs = (float(tt_dr_tm_cs) / (tt_partidas)) * 100


print("Total de partidas: {}".format(tt_partidas))
print("Number de vitórias pelo time da casa: {}".format(tt_vt_tm_cs))
print("Number de empates pelo time da casa: {}".format(tt_em_tm_cs))
print("Number de derrotas pelo time da casa: {}".format(tt_dr_tm_cs))

print("Porcentagem de vitórias pelo time da casa: {:.2f}%".format(pc_vt_cs))
print("Porcentagem de empates pelo time da casa: {:.2f}%".format(pc_em_cs))
print("Porcentagem de derrotas pelo time da casa: {:.2f}%".format(pc_dr_cs))

print("\n")

# Partidas vencidadas pelo time da casa
tt_vt_tm_vs = len(df_campeonato.query('FinalVisitante == "v"'))
tt_em_tm_vs = len(df_campeonato.query('FinalVisitante == "e"'))
tt_dr_tm_vs = len(df_campeonato.query('FinalVisitante == "d"'))

# Porcentagem de vitórias
pc_vt_vs = (float(tt_vt_tm_vs) / (tt_partidas)) * 100
pc_em_vs = (float(tt_em_tm_vs) / (tt_partidas)) * 100
pc_dr_vs = (float(tt_dr_tm_vs) / (tt_partidas)) * 100

print("Number de vitórias pelo time visitante: {}".format(tt_vt_tm_vs))
print("Number de empates pelo time da casa: {}".format(tt_em_tm_vs))
print("Number de derrotas pelo time da casa: {}".format(tt_dr_tm_vs))

print("Porcentagem de vitórias pelo time visitante: {:.2f}%".format(pc_vt_vs))
print("Porcentagem de empates pelo time visitante: {:.2f}%".format(pc_em_vs))
print("Porcentagem de derrotas pelo time visitante: {:.2f}%".format(pc_dr_vs))
```

<h3>Relação de vitórias, derrotas e empates entre o time visitante e o time da casa</h3>

```python
# Declaração de Variáveis
qt_vt_casa = df_campeonato.query('FinalCasa == "v"')['FinalCasa'].shape[0]
qt_ep_casa = df_campeonato.query('FinalCasa == "e"')['FinalCasa'].shape[0]
qt_dr_casa = df_campeonato.query('FinalCasa == "d"')['FinalCasa'].shape[0]

# Declaração de Variáveis
qt_vt_vis = df_campeonato.query('FinalVisitante == "v"')['FinalVisitante'].shape[0]
qt_ep_vis = df_campeonato.query('FinalVisitante == "e"')['FinalVisitante'].shape[0]
qt_dr_vis = df_campeonato.query('FinalVisitante == "d"')['FinalVisitante'].shape[0]

label = ['Vitória', 'Empate', 'Derrota']

# A utilização da função arange, poderia ser substituida por uma lista com duas posições ex.: [0,1]
tamanho = np.arange(3)

# Criação do plot para melhor visualização dos dados
BarVitoriaCasa = plt.bar(0
                           ,qt_vt_casa
                           ,color='#DFE9EB'
                           ,label='Vitória Casa'
                          )

BarEmpateCasa = plt.bar(1
                    ,qt_ep_casa
                    ,color='#2B5761'
                    ,label='Empate Casa'
                   )

BarDerrotaCasa = plt.bar(2
                    ,qt_dr_casa
                    ,color='#67919A'
                    ,label='Derrota'
                   )


plt.ylabel('Quantidade')
plt.title('Relação das vitórias, empates e derrotas pelos times da casa')
plt.xticks(tamanho, label)
plt.show();

######################################################################################
# Criação do plot para melhor visualização dos dados
BarVitoriaVis = plt.bar(0
                           ,qt_vt_vis
                           ,color='#DFE9EB'
                           ,label='Vitória Casa'
                          )

BarEmpateVis = plt.bar(1
                    ,qt_ep_vis
                    ,color='#2B5761'
                    ,label='Empate Casa'
                   )

BarDerrotaVis = plt.bar(2
                    ,qt_dr_vis
                    ,color='#67919A'
                    ,label='Derrota'
                   )


plt.ylabel('Quantidade')
plt.title('Relação das vitórias, empates e derrotas pelos times visitantes')
plt.xticks(tamanho, label)
plt.show();
```

<h3> Vitórias, derrotas e empates Time da casa</h3>

```python
df_campeonato.groupby(["Ano","FinalCasa",])["Ano"].size().unstack().plot.line()
plt.title("Por ano o resultado final para o time da casa")
plt.legend(['Derrota','Empate', 'Vitória'])
plt.xlabel('Anos')
plt.ylabel('Quantidade')
plt.show();
```

<h3>Vitórias, Derrotas e Empatese por ano (Time casa e Time Visitante)</h3>

```python
df_campeonato.groupby(["Ano","FinalVisitante"])["Rodada"].size().unstack().plot.bar()
plt.title("Relação de vitórias, derrotas e empates por time visitante por rodada em cada ano")
plt.legend(['Derrota do visitante','Empate', 'Vitória do visitante'])
plt.xlabel('Classe')
plt.ylabel('Quantidade')
plt.show();
```

```python
df_campeonato.groupby(["Ano","FinalCasa"])["Rodada"].size().unstack().plot.bar()
plt.title("Relação de vitórias, derrotas e empates por time visitante por rodada em casa ano")
plt.legend(['Derrota em casa','Empate', 'Vitória em casa'])
plt.xlabel('Classe')
plt.ylabel('Quantidade')
plt.show();
```

<h3>Visão de gols por ano (Time casa e Time Visitante)</h3>

```python
# Time da casa
df_campeonato.groupby(["Ano","GolCasa"])["FinalCasa"].size().unstack().plot.line()
plt.title("Gols por ano para time da casa")
plt.xlabel('Ano')
plt.ylabel('Quantidade de gols')
plt.show();
```

```python
# Time visitante
df_campeonato.groupby(["Ano","GolVisitante"])["FinalVisitante"].size().unstack().plot.line()
plt.title("Gols por ano para o time visitante")
plt.xlabel('Ano')
plt.ylabel('Quantidade de gols')
plt.show();
```

<h3>Relação de Gols (Time Visitante, Time Casa)</h3>

```python
sns.scatterplot(x='GolCasa',y='GolVisitante',data=df_campeonato)
plt.title("Relação de gols dos time da casa com o time visitante")
plt.xlabel('Gols da Casa')
plt.ylabel('Gols do Visitante')
plt.show();
```

<h3>Relação de gols (time casa | time visitante x Semana)</h3>

```python
sns.relplot(x="GolCasa", y="DiaSemana", data=df_campeonato)
plt.title("Relação de gols dos time da casa com dia da semana")
plt.xlabel('Gols da Casa')
plt.ylabel('Dias da semana')
plt.show();
```

```python
sns.relplot(x="GolVisitante", y="DiaSemana", data=df_campeonato)
plt.title("Relação de gols dos time visitante com dia da semana")
plt.xlabel('Gols Visitantes')
plt.ylabel('Dias da semana')
plt.show();
```

```python
sns.relplot(x="GolVisitante", y="Hora", data=df_campeonato)
plt.title("Relação de gols do time visitante por hora da partida")
plt.xlabel('Gols Visitantes')
plt.ylabel('Hora de Jogo')
plt.show();
```
